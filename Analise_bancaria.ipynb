```python
# Rodando o CSV transações  bancárias
import pandas as pd

df = pd.read_csv("transacoes_bancarias.csv")

df.head()
```

```python
#Estrutura Geral
#O dataset é composto por variáveis numéricas, categóricas e temporais.
#As colunas id_transacao e id_cliente são do tipo inteiro, a coluna valor é numérica contínua, data_transacao é do tipo datetime, e as colunas tipo_transacao, conta_origem, conta_destino e descricao são variáveis categóricas.
df.info()
```

```python
#Conversão de data 
df['data_transacao'] = pd.to_datetime(df['data_transacao'])
```

```python
#A coluna data_transacao foi convertida para o tipo datetime utilizando a função pd.to_datetime(), permitindo a realização de análises temporais como agrupamento por mês e ano.
df.info()
```

```python
# A análise de valores mostrou que apenas a coluna conta_destino possui valores ausentes.
df.isnull().sum()
```

```python
#A análise de valores distintos foi realizada com o método df.nunique(), permitindo identificar a diversidade de informações presentes em cada coluna do dataset.
df.nunique()
```

```python
#Tratamento de valores nulos
#No tratamento dos dados, os valores nulos identificados na coluna conta_destino foram preenchidos com o valor “Não se aplica”, pois transações do tipo saque e depósito não possuem conta de destino.
df['conta_destino'] = df['conta_destino'].fillna('Não se aplica')
```

```python
#Verificação de valores constantes
#Não foram identificadas colunas com valores constantes no dataset, indicando variabilidade adequada para análise.
df.nunique()
```

```python
# Verificação de valores inválidos
#Não foram encontrados valores negativos ou zerados na coluna valor.
df[df['valor'] <= 0]
```

```python
# Pergunta 1
#Qual o total movimentado por tipo de transação?
df.groupby('tipo_transacao')['valor'].sum()
```

```python
# Pergunta 2
#Qual o valor médio das transações por tipo?
df.groupby('tipo_transacao')['valor'].mean()
```

```python
#Pergunta 3
#Quais são os 5 clientes que mais movimentaram dinheiro?
df.groupby('id_cliente')['valor'].sum().sort_values(ascending=False).head(5)
```

```python
#Pergunta 4
#Qual o volume total movimentado por mês?

df['mes'] = df['data_transacao'].dt.to_period('M')

df.groupby('mes')['valor'].sum()
```

```python
#Pergunta 5
#Quais foram as contas de origem que mais realizaram transferências?

df[df['tipo_transacao'] == 'transferencia'] \
    .groupby('conta_origem')['valor'] \
    .sum() \
    .sort_values(ascending=False)
```
